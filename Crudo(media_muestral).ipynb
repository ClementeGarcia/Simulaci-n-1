**Mc Crudo (Media muestral)**

I = $\int_{a}^{b}g(x) = \int_{a}^{b}w(x)f(x)dx = E[w(x)] \approx \frac{1}{N}\sum_{i=1}^{N}g(x_i) $


$w(x) = g(x)(b-a)$

$f(x) = \frac{1}{b-a} ‚üπ U(a,b)$




```python
import random
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad
```

```python
def f(x):
      return np.sqrt(np.arctan(x))
```

```python
y , v= quad(f, 0, 1)

print(y)
```

```python
def crudo(n):
  suma = 0
  for i in range(n):
    x = random.random()
    suma = suma + f(x)
  return suma/n
```

```python
n1=100
L=[]
for i in range(n1):
   L.append(crudo(i+1))
```

```python
plt.axhline(y = y, color= "red", linestyle = "--")
plt.plot(L)
plt.ylim(0,1)
```
