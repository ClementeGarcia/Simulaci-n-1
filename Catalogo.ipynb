##Pseudocódigo

Generar $ U \sim \mathrm{U}(0, 1) $.

Retornar $ X = a + (b - a)U $


```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = 1
b = 5
```

```python
def Uni(n):
  #Inicializar una lista vacía para alamacener los números generados
  lista = []
  for i in range(n):
    #Se genera un número aleatorio uniforme entre 0 y 1
    R = rd.random()
    x = a + (b - a)*R
    lista.append(x)
    #Retora la lista con los números generados
  return lista
```

```python
l = Uni(1000000)
```

```python
plt.hist(l, bins=10) #Histograma de una distribución triangular
plt.title("Histograma de una Distribución Uniforme")
plt.xlabel("")
plt.ylabel("Densidad de probabilidad")
plt.grid(True, linestyle='--', alpha=0.5)
```

# Distribución triángular
# $X\sim Triangular(a,b,c)$

## Función de densidad de probabilidad

$$
f(x) =
\begin{cases}
0, & x < a \\
\frac{2(x - a)}{(b - a)(c - a)}, & a \le x < c \\
\frac{2(b - x)}{(b - a)(b - c)}, & c \le x \le b \\
0, & x > b
\end{cases}
$$

##Función de distribución acumulada

$$
F(x) =
\begin{cases}
0, & x < a \\
\frac{(x - a)^2}{(b - a)(c - a)}, & a \le x < c \\
1 - \frac{(b - x)^2}{(b - a)(b - c)}, & c \le x \le b \\
1, & x > b
\end{cases}
$$

##Pseudocódigo

Generar un número aleatorio $R \sim U(0, 1)$

Calcular $F_c =\frac{c - a}{b - a}$

Si $R < F_c$ entonces:

$X = a + \sqrt{R * (b - a) * (c - a)}$

En otro caso:

$X = b - \sqrt{(1 - R) * (b - a) * (b - c)}$

Retornar $X$

```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = 1
b = 3
c = 5
```

```python
def tri(n):
  #Inicializar una lista vacía para alamacener los números generados
  x = []
  for i in range(n):
    #Se genera un número aleatorio uniforme entre 0 y 1
    R = rd.random()
    #Si R es menor que 0.5, use la primera transforamción inversa
    if  R <0.5:
      x.append(np.sqrt(8*R)+1)
    else:
      x.append(5 - np.sqrt((1-R)*8))
      #Retora la lista con los números generados
  return x
```

```python
a = tri(100000)
```

```python
plt.hist(a, bins=200) #Histograma de una distribución triangular
plt.title("Histograma de una Distribución Triangular")
plt.xlabel("")
plt.ylabel("Densidad de probabilidad")
plt.grid(True, linestyle='--', alpha=0.5)
```
