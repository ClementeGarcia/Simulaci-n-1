###Función de distribución acumulada

$$
F(x) = \begin{cases}
0 & \text{si } x < a \\[0.5em]
\frac{x - a}{b - a} & \text{si } a \leq x \leq b \\[0.5em]
1 & \text{si } x > b
\end{cases}
$$

$\textbf{Media:}$

$$
