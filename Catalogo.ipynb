##Pseudocódigo

Generar $ U \sim \mathrm{U}(0, 1) $.

Retornar $ X = a + (b - a)U $


```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = 1
b = 5
```

```python
def Uni(n):
  #Inicializar una lista vacía para alamacener los números generados
  lista = []
  for i in range(n):
    #Se genera un número aleatorio uniforme entre 0 y 1
    R = rd.random()
    x = a + (b - a)*R
    lista.append(x)
    #Retora la lista con los números generados
  return lista
```

```python
l = Uni(1000000)
```

```python
plt.hist(l, bins=10) #Histograma de una distribución triangular
plt.title("Histograma de una Distribución Uniforme")
plt.xlabel("")
plt.ylabel("Densidad de probabilidad")
plt.grid(True, linestyle='--', alpha=0.5)
```

# Distribución triángular
# $X\sim Triangular(a,b,c)$

## Función de densidad de probabilidad

$$
f(x) =
\begin{cases}
0, & x < a \\
\frac{2(x - a)}{(b - a)(c - a)}, & a \le x < c \\
\frac{2(b - x)}{(b - a)(b - c)}, & c \le x \le b \\
0, & x > b
\end{cases}
$$

image.png

##Función de distribución acumulada

$$
F(x) =
\begin{cases}
0, & x < a \\
\frac{(x - a)^2}{(b - a)(c - a)}, & a \le x < c \\
1 - \frac{(b - x)^2}{(b - a)(b - c)}, & c \le x \le b \\
1, & x > b
\end{cases}
$$

image.png

$$$$
$\textbf{Media:}$

$$
E[X] = \frac{a + b + c}{3}
$$

$\textbf{Varianza:}$

$$
Var(X) = \frac{a^2 + b^2 + c^2 - ab - ac - bc}{18}
$$
$\textbf{Desviación estándar:}$
$$
\sigma = \sqrt{Var(X)}
$$

$\textbf{Parámetros:}$

$$
a \leq c \leq b
$$


##Pseudocódigo

Generar un número aleatorio $R \sim U(0, 1)$

Calcular $F_c =\frac{b - a}{c - a}$

Si $R < F_c$ entonces:

$X = a + \sqrt{R * (b - a) * (c - a)}$

En otro caso:

$X = c - \sqrt{(1 - R) * (c - b) * (c - a)}$

Retornar $X$

```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = 1
b = 3
c = 5
```

```python
def tri(n):
  #Inicializar una lista vacía para alamacener los números generados
  x = []
  for i in range(n):
    #Se genera un número aleatorio uniforme entre 0 y 1
    R = rd.random()
    #Si R es menor que 0.5, use la primera transforamción inversa
    if  R <(b-a)/(c-a):
      x.append(a + np.sqrt(R * (b - a) * (c - a)))
    else:
      x.append(c - np.sqrt((1 - R) * (c - b) * (c - a)))
      #Retora la lista con los números generados
  return x
```

```python
k = tri(100000)
```

```python
plt.hist(k, bins=200) #Histograma de una distribución triangular
plt.title("Histograma de una Distribución Triangular")
plt.xlabel("")
plt.ylabel("Densidad de probabilidad")
plt.grid(True, linestyle='--', alpha=0.5)
```

#Distribución Exponencial $X \sim Exp(\lambda)$

## Función de densidad de probabilidad

$$
f(x) =
\begin{cases}
\lambda e^{-\lambda x}, & x \geq 0 \\
0, & x < 0
\end{cases}
$$

image.png

## Función de distribución acumulada

$$
F(x) =
\begin{cases}
1 - e^{-\lambda x}, & x \geq 0 \\
0, & x < 0
\end{cases}
$$

image.png

$\textbf{Media:}$

$$
E[X] = \frac{1}{\lambda}
$$

$\textbf{Varianza:}$

$$
Var(X) = \frac{1}{\lambda^2}
$$
$\textbf{Desviación estándar:}$
$$
\sigma = \sqrt{Var(X)} = \frac{1}{\lambda}

$\textbf{Parámetro:}$

$$
\lambda >0
$$

##Pseudocódigo

Generar un número aleatorio R ~ U(0, 1)

Calcular $X = -\frac {1}{λ} * ln(1 - R)$

```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np
```

```python
lmba = 1.5
```

```python
def exp(n):
    lista = []
    for i in range(n):
        R = rd.random()
        x = -np.log(R) / lmba
        lista.append(x)
    return lista
```

# Distribución m-Erlang
# $X \sim Erlang(m,\lambda)$

## Función de densidad de probabilidad

$$
f(x) =
\begin{cases}
\dfrac{\lambda^m x^{m - 1} e^{-\lambda x}}{(m - 1)!}, & x \geq 0 \\
0, & x < 0
\end{cases}
$$

image.png

## Función de densidad acumulada

$$
F(x) =
\begin{cases}
1 - \sum\limits_{k=0}^{m-1} \dfrac{(\lambda x)^k e^{-\lambda x}}{k!}, & x \geq 0 \\
0, & x < 0
\end{cases}
$$

image.png

$\textbf{Media:}$

$$
E[X] = \frac{m}{\lambda}
$$

$\textbf{Varianza:}$

$$
Var(X) = \frac{m}{\lambda^2}
$$

$\textbf{Desviación estándar:}$
$$
\sigma = \sqrt{Var(X)} = \frac{\sqrt{m}}{\lambda}
$$

$\textbf{Parámetros:}$

$$
m \in Z^+,   \lambda>0
$$

## Pseudocódigo

Generar $ U_1, U_2, \ldots, U_m $ $\sim U(0, 1)$.

Retornar $ X = x +\frac{-\beta}{m} \ln \left( \prod_{i=1}^m U_i \right) $. , si $\beta = \frac{m}{\lambda}$ entonces $X \sim Erlang(m,\lambda)$


```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np
```

```python
m = 4
labda = 2
```

```python
def erlang(n):
    lista = []
    for i in range(n):
        x = 0
        for i in range(m):
            R = rd.random()
            x += -np.log(R) / labda
        lista.append(x)
    return lista
```

```python
l = erlang(1000000)
```

```python
plt.hist(l, bins=2000) #Histograma de una distribución erlang
plt.title("Histograma de una Distribución Erlang")
plt.xlabel("")
plt.ylabel("Densidad de probabilidad")
plt.grid(True, linestyle='--', alpha=0.5)
```

# Distribución Gamma $X \sim \Gamma (\alpha, \lambda)$

##Función de densidad de probabilidad

$$
f(x) =
\begin{cases}
\dfrac{\lambda^\alpha x^{\alpha - 1} e^{-\lambda x}}{\Gamma(\alpha)}, & x \geq 0 \\
0, & x < 0
\end{cases}
$$

image.png
##Función de distribución acumulada


$$
F(x) =
\begin{cases}
\dfrac{1}{\Gamma(\alpha)} \int_0^{\lambda x} t^{\alpha - 1} e^{-t} dt = \dfrac{\gamma(\alpha, \lambda x)}{\Gamma(\alpha)}, & x \geq 0 \\
0, & x < 0
\end{cases}
$$

image.png

$\textbf{Media:}$

$$
E[X] = \frac{\alpha}{\lambda}
$$

$\textbf{Varianza:}$

$$
Var(X) = \frac{\alpha}{\lambda^2}
$$

$\textbf{Desviación estándar:}$
$$
\sigma = \sqrt{Var(X)} = \frac{\sqrt{\alpha}}{\lambda}
$$

$\textbf{Parámetros:}$

$$
\alpha >0 , \lambda >0
$$

## Pseudocódigo

Inicializar $X = 0$

Para i = 1 hasta α hacer:

Generar $R \sim U(0, 1)$
    
Calcular $E = -\frac{ln(R)}{λ}$
    
Acumular: $X ← X + E$

Retornar $X$





```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np
```

```python
alpha = 6
labda = 1
```

```python
def gamma(n):
    lista = []
    for i in range(n):
        x = 0
        for i in range(alpha):
            R = rd.random()
            x += -np.log(R) / labda
        lista.append(x)
    return lista
```

```python
p = gamma(1000000)
```

```python
plt.hist(p, bins=2000) #Histograma de una distribución gamma
plt.title("Histograma de una Distribución Gamma")
plt.xlabel("")
plt.ylabel("Densidad de probabilidad")
plt.grid(True, linestyle='--', alpha=0.5)
```

#Distribución Weibull
$X \sim \text{Weibull}(k,\lambda)
$

## Función de densidad de probabilidad

$$
f(x) =
\begin{cases}
\frac{k}{\lambda} \left( \frac{x}{\lambda} \right)^{k - 1} e^{-(x / \lambda)^k}, & x \geq 0 \\
0, & x < 0
\end{cases}
$$

image.png

## Función de distribución acumulada

$$
F(x) =
\begin{cases}
1 - e^{-(x / \lambda)^k}, & x \geq 0 \\
0, & x < 0
\end{cases}
$$

image.png

$\textbf{Media:}$

$$
E[X] = \lambda \, \Gamma\left(1 + \frac{1}{k}\right)
$$

$\textbf{Varianza:}$

$$
Var(X) = \lambda^2 \left[ \Gamma\left(1 + \frac{2}{k}\right) - \left(\Gamma\left(1 + \frac{1}{k}\right)\right)^2 \right]
$$

$\textbf{Desviación estándar:}$
$$
\sigma = \sqrt{Var(X)}
$$

$\textbf{Parámetros:}$

$$
k >0 , \lambda >0
$$

## Pseudocódigo

Generar $R \sim U(0, 1)$

Calcular $X = \lambda * [ -ln(1 - R) ]^{\frac{1}{k}}$

Retornar $X$


```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np
```

```python
k = 1.5
labda = 2
```

```python
def weibull(n):
    lista = []
    for i in range(n):
        R = rd.random()
        x = labda * (-np.log(R))**(1 / k)
        lista.append(x)
    return lista
```

# Distribución Normal
$$
X \sim \mathcal{N}(\mu, \sigma^2)
$$

## Función de densidad de probabilidad

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right), \quad x \in \mathbb{R}
$$
image.png

## Función de distribución acumulada

Función error:

$$
F(x) = \frac{1}{2} \left[ 1 + \text{erf}\left( \frac{x - \mu}{\sigma \sqrt{2}} \right) \right]
$$
image.png

$\textbf{Media:}$

$$
E[X] = \mu
$$

$\textbf{Varianza:}$

$$
Var(X) = \sigma^2
$$

$\textbf{Desviación estándar:}$
$$
\sigma = \sqrt{Var(X)}
$$

$\textbf{Parámetros:}$

$$
\mu \in \Re , \sigma >0
$$

## Pseudocódigo

 Generar  $R_1$ y $R_2$ $ \sim U(0, 1)$

Calcular:
    
$Z = \sqrt{-2 * ln(R_1)} * cos(2 * \pi * R_2)$

Transformar:
$X = \mu + \sigma * Z$

Retornar $X$

```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np
```

```python
mu = 4
sigma = 1.7
```

```python
def normal(n):
    lista = []
    for i in range(n):
        R1 = rd.random()
        R2 = rd.random()
        Z = np.sqrt(-2 * np.log(R1)) * np.cos(2 * np.pi * R2)
        X = mu + sigma * Z
        lista.append(X)
    return lista
```

```python
o = normal(1000000)
```

```python
plt.hist(o, bins=2000) #Histograma de una distribución normal
plt.title("Histograma de una Distribución Normal")
plt.xlabel("")
plt.ylabel("Densidad de probabilidad")
plt.grid(True, linestyle='--', alpha=0.5)
```

# Distribución Log-normal
$$X \sim \text{LogNormal}(\mu, \sigma^2)$$
donde:
$$Y = \ln(X) \sim \mathcal{N}(\mu, \sigma^2)$$

# Función de densidad de probabilidad

$$
f(x) =
\begin{cases}
\frac{1}{x \sigma \sqrt{2\pi}} \exp\left( -\frac{(\ln x - \mu)^2}{2\sigma^2} \right), & x > 0 \\
0, & x \leq 0
\end{cases}
$$

image.png

## Función de distribución acumulada

$$
F(x) =
\begin{cases}
\frac{1}{2} \left[ 1 + \text{erf} \left( \frac{\ln x - \mu}{\sigma \sqrt{2}} \right) \right], & x > 0 \\
0, & x \leq 0
\end{cases}
$$

image.png

$\textbf{Media:}$

$$
E[X] = e^{\mu + \frac{\sigma^2}{2}}
$$

$\textbf{Varianza:}$

$$
Var(X) = \left( e^{\sigma^2} - 1 \right) e^{2\mu + \sigma^2}
$$

$\textbf{Desviación estándar:}$
$$
\sigma = \sqrt{Var(X)} = e^{\mu + \frac{\sigma^2}{2}}* \sqrt{e^{\sigma^2} - 1}
$$

$\textbf{Parámetros:}$

$$
X >0 , \mu \in \mathbb{R}, \sigma >0
$$

## Pseudocódigo

Generar $R_1, R_2 \sim U(0, 1)$

Calcular:
$$Z = \sqrt{-2 * ln(U_1)} * cos(2 * \pi * U_2)$$

Obtener $Y = \mu + \sigma * Z$

Calcular $X = exp(Y)$

Retornar $X$

```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np
```

```python
mu = 1
sigma = 0.8

```

```python
def lognormal(n):
    lista = []
    for i in range(n):
        R1 = rd.random()
        R2 = rd.random()
        Z = np.sqrt(-2 * np.log(R1)) * np.cos(2 * np.pi * R2)
        Y = mu + sigma * Z
        x = np.exp(Y)
        lista.append(x)
    return lista
```

```python
k = lognormal(1000000)
```

```python
plt.hist(k, bins=2000) #Histograma de una distribución lognormal
plt.title("Histograma de una Distribución Lognormal")
plt.xlabel("")
plt.ylabel("Densidad de probabilidad")
plt.grid(True, linestyle='--', alpha=0.5)
```

#Distribución Beta

$$
X \sim \text{Beta}(\alpha, \beta)
$$

## Función de densidad de probabilidad

$$
f(x) =
\begin{cases}
\dfrac{1}{B(\alpha, \beta)} x^{\alpha - 1} (1 - x)^{\beta - 1}, & 0 < x < 1 \\
0, & \text{en otro caso}
\end{cases}
$$
Donde:  

$$
B(\alpha, \beta) = \int_0^1 t^{\alpha - 1}(1 - t)^{\beta - 1} dt = \dfrac{\Gamma(\alpha) \Gamma(\beta)}{\Gamma(\alpha + \beta)}
$$

image.png

## Función de distribución acumulada

$$
F(x) = \dfrac{1}{B(\alpha, \beta)} \int_0^x t^{\alpha - 1} (1 - t)^{\beta - 1} dt = I_x(\alpha, \beta)
$$

image.png


$\textbf{Media:}$

$$
E[X] = \frac{\alpha}{\alpha + \beta}
$$

$\textbf{Varianza:}$

$$
Var(X) = \frac{\alpha \beta}{(\alpha + \beta)^2 (\alpha + \beta + 1)}
$$

$\textbf{Desviación estándar:}$
$$
\sigma = \sqrt{Var(X)} = \sqrt{ \frac{\alpha \beta}{(\alpha + \beta)^2 (\alpha + \beta + 1)}}
$$

$\textbf{Parámetros:}$

$$
\alpha > 0, \ \beta > 0
$$

## Pseudocódigo

Generar $G_1 \sim \Gamma(\alpha, 1)$         # con tasa $\lambda = 1$

Generar $G_2 \sim \Gamma(\beta, 1)$

Calcular: X = G1 / (G1 + G2)

Retornar X

```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np
```

```python
alpa = 2.5
betha = 3.5
```

```python
def beta(n):
    lista = []
    for i in range(n):
        # Gamma por método de suma de exponenciales
        G1 = sum(-np.log(rd.random()) for i in range(int(alpa)))
        G2 = sum(-np.log(rd.random()) for i in range(int(betha)))
        X = G1 / (G1 + G2)
        lista.append(X)
    return lista
```

```python
g = beta(1000000)
```

```python
plt.hist(g, bins=2000) #Histograma de una distribución beta
plt.title("Histograma de una Distribución Beta")
plt.xlabel("")
plt.ylabel("Densidad de probabilidad")
plt.grid(True, linestyle='--', alpha=0.5)
```

$$
\frac{1}{f(x)} \frac{df(x)}{dx} = -\frac{a_1 + a_2 x}{1 + b_1 x + b_2 x^2}
$$

Donde $a_1, a_2, b_0, b_1, b_2 $ son constantes que determinan el tipo específico de distribución de Pearson.

## Familia de Pearson

1. Tipo 0 Normal
2. Tipo I Bounded(asimétrica)
3. Tipo II simétrica, acotada
4. Tipo III Solo un límite (positiva, asimétrica)
5. Tipo IV Cola pesada(Unilateral)
6. Dos parámetros, distribuciones pesadas
7. Tipo VII Simétrica con colas pesadas

## Función de distribución acumulada

$$
F(x) = \int_{-\infty}^{x} f(t) \, dt
$$

$\textbf{Media:}$

$$
E[X] = \mu
$$

$\textbf{Varianza:}$

$$
Var(X) = \mathbb{E}[X^2] - \mu^2
$$

$\textbf{Desviación estándar:}$
$$
\sigma = \sqrt{Var(X)}
$$
