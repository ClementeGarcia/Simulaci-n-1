```python
import tabulate
import matplotlib.pyplot as plt
import numpy as np
import random
def generador(a,c,m,x0):
    x1=(a*x0+c) %m
    return x1

n=10000
a=101
c=21
m=100
x0=1

muestra=[]
datos=[]

for i in range(n):
    nuevo=generador(a,c,m,x0)
    muestra.append([i,x0,nuevo/m])
    datos.append(x0/m)
    x0=nuevo

encabezados=["i", "x_i", "X_i/m"]
print(tabulate.tabulate(muestra, headers=encabezados,tablefmt="psql",numalign="center"))

print(datos)


plt.hist(datos, bins = 10, color = "blue", rwidth=0.9)
plt.title("Ejemplo Histograma")
plt.xlabel("Nmeros Aleatorios")
plt.ylabel("Frecuencia")
plt.show()



```
