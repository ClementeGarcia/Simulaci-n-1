```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(350, 2000, 400)

y1 = (60000 - 40 * x) / 25  # 40x + 25y <= 60000 -> y <= (60000 - 40x)/25
y_min = 400 * np.ones_like(x)  # y >= 400
y_max = 1000 * np.ones_like(x)  # y <= 1000
x_min = 375  # x >= 375

plt.figure(figsize=(10, 6))
plt.plot(x, y1, label=r'$40x + 25y \leq 60000$', color='blue')
plt.axvline(x=375, label=r'$x \geq 375$', color='green')
plt.axhline(y=400, label=r'$y \geq 400$', color='red')
plt.axhline(y=1000, label=r'$y \leq 1000$', color='orange',)

# Región factible
x_vals = np.linspace(375, 1250, 400)
y_upper = np.minimum((60000 - 40 * x_vals) / 25, 1000)
y_lower = np.maximum(400, 0)
plt.fill_between(x_vals, y_lower, y_upper, where=(y_upper > y_lower), color='lightgray', alpha=0.5)
puntos_extremos = [(375, 400), (375, 1000), (875, 1000), (1250, 400)]
for punto in puntos_extremos:
    plt.scatter(*punto, color='purple', zorder=5)
    plt.text(punto[0], punto[1], f' ({punto[0]}, {punto[1]})', fontsize=9)

plt.xlim(350, 2000)
plt.ylim(300, 2000)
plt.xlabel('Eastern Cable')
plt.ylabel('ComSwitch')
plt.title('Región factible ')
plt.legend()
plt.grid(True)
plt.show()
```
