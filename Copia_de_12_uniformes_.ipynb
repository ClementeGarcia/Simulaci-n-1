#Método de las 12 uniformes

 Método de las 12 Uniformes también conocido como Método de las 12 Uniformes para Generar una Distribución Normal, es una técnica estadística utilizada para generar números aleatorios que sigan una distribución normal estándar media = 0, desviación estándar = 1 a partir de la suma de variables aleatorias uniformes.

 Este método se basa en el Teorema del Límite Central, que establece que la suma de un gran número de variables aleatorias independientes e idénticamente distribuidas tiende a una distribución normal.

 Para generar:

Se generan numeros aleatorios normales
 $$
f_X(x) = \frac{1}{\sqrt{2\pi\sigma^2}}\, \exp\left[-\frac{(x-\mu)^2}{2\sigma^2}\right],\quad -\infty < x < \infty.
$$

Generamos "n" números aleatorios y consideramos

$$ z = \frac{Y - \sum_{i=1}^{n} a_i \mu_i}{\sqrt{\sum_{i=1}^{n} a_i^2 \sigma_i^2}} $$
Esta fórmula estandariza una variable $Y$ restando su media y dividiendo por su desviación estándar


donde :

$$y = R_1 + R_2 + \cdots + R_n $$

Representa la suma de n variables aleatorias, por el teorema del mímite central, si n es grande, $y$ tenderá a una distribución normal.
$\$

Indica que $h_i$ sigue una distribución uniforme en $[0,1]$

$$ h_i \sim U(0,1) $$

Este es el método clásico de las 12 uniformes para generar normales estándar

$$z = \sum_{i=1}^{n} R_i - 6 $$

Transformacióm a una normal $N(\mu, \sigma^2)$

$$x = \mu + \sigma \left( \sum_{i=1}^{n} R_i - 6 \right) $$

```python
import numpy as np
import matplotlib.pyplot as plt
from random import random
```

```python
mu = 2 #Con media de 2
sigma = 3 #Con una sigma de 3
```

```python
def espe(n):
  lista = []   #lista vacía
  for _ in range(n):
    suma = 0
    for _ in range(12): # Sumar 12 números aleatorios U(0,1)
      R = random()      # Genera un número aleatorio en [0,1)
      suma += R
    x = mu + sigma * (suma -6)   # Transformar a N(mu, sigma^2)
    lista.append(x)

  return lista
```

```python
a = espe(100000) # Genera 100,000 números aleatorios normales
```

```python
plt.hist(a, bins = 50)
plt.title('Distribución ', fontsize=14)
plt.xlabel('Valores', fontsize=12)
plt.ylabel('Densidad de Probabilidad', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
```
